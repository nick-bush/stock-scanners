```python
df_new = pd.DataFrame(btc['Adj Close'])
```

```python
type(df_new)
```

```python
df_new.rename(columns={'Adj Close':'BTC'},inplace=True)
df_new
```

```python
df_new['TSLA'] = tsla['Adj Close']
df_new['SPY'] = spy['Adj Close']
df_new['ETH'] = eth['Adj Close']
```

```python
df_new
```

```python

for i in df_new.columns[1:]:
    for j in range(0,len(df_new)):
        #must use isnan to compare nan values
        if np.isnan(df_new[i][j]):
            #print('FOUND ONE')
            #print('Old: {}'.format(df_new[i][j]))
            df_new[i][j] = df_new[i][j-1]
            #print('New: {}'.format(df_new[i][j]))
            
```

```python
df_new
```

```python
def daily_return(df):
    df_daily_return = df.copy()
    #loop on columns
    for i in df.columns:
        #Loop on rows
        for j in range(1,len(df)):
            df_daily_return[i][j] = ( (df[i][j] -df[i][j-1]) /df[i][j-1] )*100
        df_daily_return[i][0] = df_daily_return[i].mean()
    return df_daily_return
```

```python
df_returns = daily_return(df_new)
```

```python
df_returns
```
