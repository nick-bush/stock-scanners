```python
import yfinance as yf
from yahoo_fin import stock_info as si
from pandas_datareader import data as pdr
```

```python
hourly_data = yf.download('TSLA', period="365d", interval='1d')
```

```python
hourly_data['34 EMA'] = hourly_data['Adj Close'].ewm(min_periods=34, adjust=False, alpha =(2/35)).mean().round(2)
hourly_data['50 EMA'] = hourly_data['Adj Close'].ewm(min_periods=50,adjust=False,alpha=(2/50)).mean().round(2)
```

```python
df = hourly_data.tail(315)
```
