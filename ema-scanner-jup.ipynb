```python
#this code currently returning incorrect EMA's relative to trading platforms
hourly_data['34 EMA'] = hourly_data['Adj Close'].ewm(span=34, adjust=False).mean().round(2)
hourly_data['50 EMA'] = hourly_data['Adj Close'].ewm(span=50).mean().round(2)
```
